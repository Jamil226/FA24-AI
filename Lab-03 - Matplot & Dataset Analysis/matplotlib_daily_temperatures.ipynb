## **Matplotlib**

Matplotlib is a powerful library for creating static, animated, and interactive visualizations in Python. Here are some key features and functionalities:

- **Plot Types**: Matplotlib supports various plot types such as line plots, bar charts, histograms, scatter plots, and more.
- **Customization**: You can customize almost every aspect of a plot, including colors, markers, line styles, axes, titles, and labels.
- **Subplots**: You can create multiple plots in a single figure, allowing for complex visualizations.
- **Interactivity**: Matplotlib supports interactive plots, enabling users to zoom, pan, and save figures.


```python
import matplotlib.pyplot as plt
import numpy as np
```

## **Line Plot of Daily Temperatures**

```python
temperatures = np.array([36, 39, 34, 42, 40, 38, 33])
days = np.arange(1, len(temperatures) + 1)  # Days of the week

average_temp = np.mean(temperatures)

# Line Plot
plt.figure(figsize=(10, 5))
plt.plot(days, temperatures, marker='o', linestyle='-', color='blue', label='Daily Temperature')
plt.axhline(y=average_temp, color='r', linestyle='--', label='Average Temperature')
plt.title('Daily Temperatures Over the Week')
plt.xlabel('Days of the Week')
plt.ylabel('Temperature (C)')
plt.xticks(days)
plt.legend()
plt.grid()
plt.show()
```

## **Bar Chart of Daily Temperatures**

```python
# Bar Chart
plt.figure(figsize=(10, 5))
plt.bar(days, temperatures, color='b', label='Daily Temperature')
plt.axhline(y=average_temp, color='r', linestyle='--', label='Average Temperature')
plt.title('Daily Temperatures Over the Week')
plt.xlabel('Days of the Week')
plt.ylabel('Temperature (C)')
plt.xticks(days)
plt.legend()
plt.show()
```

## **Scatter Plot of Daily Temperatures**

```python
plt.figure(figsize=(10, 5))
plt.scatter(days, temperatures, color='b', label='Daily Temperature', s=100)
plt.axhline(y=average_temp, color='r', linestyle='--', label='Average Temperature')
plt.title('Daily Temperatures Over the Week')
plt.xlabel('Days of the Week')
plt.ylabel('Temperature (C)')
plt.xticks(days)
plt.legend()
plt.grid()
plt.show()
```

## **Histogram**

```python
# Histogram
plt.figure(figsize=(10, 5))
plt.hist(temperatures, bins=5, color='g', alpha=0.7)
plt.title('Temperature Distribution (Histogram)')
plt.xlabel('Temperature (C)')
plt.ylabel('Frequency')
plt.axvline(x=np.mean(temperatures), color='r', linestyle='--', label='Average Temperature')
plt.legend()
plt.grid()
plt.show()
```

## **Pie Chart**

```python
# Pie Chart - Example categorization
temperature_bins = [0, 30, 35, 40, 45]  # Define bins
labels = ['< 30C', '30-35C', '35-40C', '40-45C']
hist, _ = np.histogram(temperatures, bins=temperature_bins)

plt.figure(figsize=(8, 8))
plt.pie(hist, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Temperature Range Distribution (Pie Chart)')
plt.show()
```
